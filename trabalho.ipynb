```python
# prompt: codigo para pesquisar por coluna

import pandas as pd
from google.colab import auth
import gspread
from google.auth import default

auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Substitua 'Nome da sua planilha' pelo nome da sua planilha do Google Sheets
worksheet = gc.open('trabalho').sheet1

# Obter todos os valores da planilha
rows = worksheet.get_all_values()

# Criar um DataFrame do Pandas com os dados
df_manutencoes = pd.DataFrame.from_records(rows[1:], columns=rows[0])

def pesquisar_manutencoes(df):
  coluna_pesquisa = input("Digite o nome da coluna para pesquisa: ")
  termo_pesquisa = input(f"Digite o termo de pesquisa na coluna '{coluna_pesquisa}': ")

  if coluna_pesquisa not in df.columns:
    print(f"Coluna '{coluna_pesquisa}' não encontrada no DataFrame.")
    return

  # Realiza a pesquisa na coluna especificada, tratando possíveis erros
  try:
    resultados = df[df[coluna_pesquisa].astype(str).str.contains(termo_pesquisa, case=False, na=False)]
    print("\nResultados da pesquisa:")
    print(resultados)
  except KeyError:
    print(f"Erro: Coluna '{coluna_pesquisa}' não encontrada ou inválida.")
  except AttributeError:
      print(f"Erro: Tipo de dado na coluna '{coluna_pesquisa}' não suportado para pesquisa de texto.")


while True:
  print("\nOpções:")
  print("1. Pesquisar manutenções")
  print("2. Sair")

  opcao = input("Escolha uma opção: ")

  if opcao == '1':
    pesquisar_manutencoes(df_manutencoes)
  elif opcao == '2':
    break
  else:
    print("Opção inválida.")
```
